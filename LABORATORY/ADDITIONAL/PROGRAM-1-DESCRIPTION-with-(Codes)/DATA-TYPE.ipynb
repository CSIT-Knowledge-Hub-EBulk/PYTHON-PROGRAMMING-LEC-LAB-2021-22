# Numeric Type / Number
INTEGER- DATA-TYPE

```python
#20, 10, -50, -1002
y = 10
pin_code = 564512
print(y)
print(pin_code)
```

```python
print(type(pin_code))
print(type(y))
```

#  Numeric Type / Number
FLOAT- DATA-TYPE

```python
#25.56, 10.5, -45.69, -0.8
price = 25.56
run_rate = -0.8
value = 5.1e5
run_rate = -0.8
print(price)
print(run_rate)
print(value)
```

```python
print(type(run_rate))
print(type(price))
print(type(value))
```

# # # Numeric Type / Number
