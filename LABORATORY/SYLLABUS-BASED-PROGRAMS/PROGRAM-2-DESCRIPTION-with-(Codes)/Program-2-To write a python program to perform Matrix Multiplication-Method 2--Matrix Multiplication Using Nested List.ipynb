# 2. To write a python program to perform Matrix Multiplication.
Method 2: Matrix Multiplication Using Nested List.
 We use zip in Python.

```python
# Program to multiply two matrices using list comprehension

# take a 3x3 matrix
A = [[12, 7, 3],
	[4, 5, 6],
	[7, 8, 9]]

# take a 3x4 matrix
B = [[5, 8, 1, 2],
	[6, 7, 3, 0],
	[4, 5, 9, 1]]

# result will be 3x4
result = [[sum(a * b for a, b in zip(A_row, B_col))
						for B_col in zip(*B)]
								for A_row in A]
```

```python
for r in result:
	print(r)
```
